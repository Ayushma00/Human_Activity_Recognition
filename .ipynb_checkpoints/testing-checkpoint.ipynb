```python
from keras.models import load_model
model=load_model("test2.h5")# loading the model that was previously saved in h5 format
import cv2 
import numpy as np
```

```python
input="call1.webm" # loading the unseen video for prediction
```

# Function to convert the video into frames of sequetial image arrays

```python
image_copy=np.zeros((80,80))
def get_frame(vidcap,seconds):
    global image_copy
    vidcap.set(cv2.CAP_PROP_POS_MSEC,seconds*1000)
    hasFrames,image = vidcap.read() 
    px=80 # dimension of image height
    py=80 # width
    
    if hasFrames:
        gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image1 = cv2.resize(np.uint8(gray),(px,py), interpolation = cv2.INTER_AREA)
        image_array = np.reshape(np.array(image1),(px,py))
        image_copy=image_array
    else:
        image_array=image_copy
    return hasFrames,image_array.astype("int") # returning the boolen value of having frames or not and arrays of images from the video
   
    
    
```

```python

video_cap=cv2.VideoCapture(input)
sec = 0
sec1=0
frameRate = 0.3
count=1
list2=[]
success,_ = get_frame(video_cap,sec)
while success:
    list1=[]
    count = count + 1
    sec = sec + frameRate
    sec = round(sec, 2)
    sec1 = sec1 + frameRate
    sec1 = round(sec1, 2)
#     get frames of video up to 12 seconds to maintain the time consistency of frames for each video
    if sec<=12:
        success,img =get_frame(video_cap,sec1)
        list1.append(img)
        list2.append(list1)
#         re get the frames from video if the sample is less than 12 seconds like getting the frames of 3 sec video by running 3 times.
        if success==0:
            sec1=0
            success=1
    else:
        success=0
```

```python
X1=np.array(list2)
X1.shape 
# here 40 is the total number of images formed from the video and 80*80 are dimensions
```
